```python
from random import randint
com_guess=randint(1,50)
chance=1
while chance<=5:
    user_guess=int(input('Enter The number betweeen 1 and 50'))
    if user_guess>50:
        print('Number should between 1 and 50')
        continue
    elif user_guess<com_guess:
        print('Think Big')
    elif user_guess>com_guess:
        print('Think Small')
    elif user_guess==user_guess:
        print('You won the game')
        break
    chance+=1
else:
    print('You are such looser')
    print(f'Computer Guess was {com_guess}')
```
